## Projection Demo

This notebook demonstrates the projections used in the Benchmarker code. The purpose of the Benchmarker code is to
compare the different alignment methods using different image matching algorithms. Right now we are using projections
to show the effectiveness of the alignment methods. We use pairs of images to project LIDAR data associated with one image
onto the other one. The projection acts as the ground truth for the matches returned by the OpenCV image matching algorithms.
The projection helps us visualize what one image looks like with respect to the other in space by projecting the LIDAR points 
from one into the other. This projection is done using the phone poses associated with each image. 

Refer to each file header to learn about the file in the benchmarkers directory.

```python
from Benchmarker import Benchmarker
from benchmarkers.CloudBenchmark import cloud_anchor_pose_test
from benchmarkers.PairBenchmark import benchmark
from MatchingAlgorithm import OrbMatcher, SiftMatcher, AkazeMatcher

# different algorithms to loop through 
algorithms = [OrbMatcher()]
values = [0.5]

benchmarker = Benchmarker(algorithms, values)
```

### Cloud Benchmarker demo

```python
cloud_anchor_pose_test(benchmarker, num_runs=1)
```

### Pair Benchmarker demo
This one if more accurate than the cloud benchmarker since it uses to consecutive phone poses in the same session to calculate the pose difference to project the LIDAR points. The cloud benchmarker currently uses the pose difference between the poses associated with the first two matching cloud anchors from 2 different sessions. Those are usually 5-10 seconds into the session and the phone has moved significantly by then.

```python
benchmark(benchmarker, num_runs=1)
```
